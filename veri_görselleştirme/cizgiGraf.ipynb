## Çizgi Grafik

```python
import seaborn as sns
import pandas as pd
fmri = sns.load_dataset("fmri")
df = fmri.copy()
df.head()
```

```python
df.shape
```

```python
df["timepoint"].describe()
```

```python
df["signal"].describe()
```

```python
df.groupby("timepoint")["signal"].count()
# her bir zaman noktasına göre eşit sayıda sayma işlemi yapılmıştır.
```

```python
df.groupby("signal").count()
# herbirisinin tekil olduğuna eriştik
```

```python
df.groupby("timepoint")["signal"].describe()
```

```python
sns.lineplot(x = "timepoint", y ="signal", data = df);
```

```python
sns.lineplot(x = "timepoint", y ="signal",hue = "event", data = df);
```

```python
sns.lineplot(x = "timepoint", y ="signal",hue = "event",style="event", data = df);
```

```python
sns.lineplot(x = "timepoint",
             y ="signal",
             hue = "event",
             style="event",
             markers =  True,
             dashes = False,
             data = df);
# ortalama değerleri işaretleyip sunar
```

```python
sns.lineplot(x = "timepoint",
             y ="signal",
             hue = "region",
             style="event",
             data = df);
```
