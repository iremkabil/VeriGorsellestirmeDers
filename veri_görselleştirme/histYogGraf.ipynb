```python
import seaborn as sns
diamonds = sns.load_dataset("diamonds")
df = diamonds.copy()
df.head()
```

```python
sns.displot(df.price);
```

```python
sns.histplot(df.price, kde=True);
# kde : yoÄŸunluk fonksiyonunun da grafik Ã¼zerine koyulup koyulmamasÄ±
```

```python
(sns.FacetGrid(df,
              hue = "cut",
              height = 5,
              xlim = (0,1000)));
```

* sns.FacetGrid: Bu, Seaborn'un FacetGrid sÄ±nÄ±fÄ±nÄ± kullanarak Ã§oklu grafikleri dÃ¼zenlemek iÃ§in kullanÄ±lan bir yapÄ±dÄ±r. FarklÄ± kategorilere gÃ¶re veriyi bÃ¶lÃ¼p her birini ayrÄ± bir grafikte gÃ¶sterebilir veya aynÄ± grafikte farklÄ± renklerle temsil edebilir.
* df: Bu, Ã¼zerinde Ã§alÄ±ÅŸtÄ±ÄŸÄ±mÄ±z Pandas DataFrame'dir. DataFrame, price ve cut sÃ¼tunlarÄ±nÄ± iÃ§erir.
* hue="cut": Bu parametre, grafiÄŸi renklerle farklÄ± kategorilere (bu durumda "cut" kategorilerine) ayÄ±rmak iÃ§in kullanÄ±lÄ±r. FarklÄ± cut kategorileri farklÄ± renklerle temsil edilecektir.
* height=5: Bu, her bir facet'in (grafik) yÃ¼ksekliÄŸini belirler. Grafik 5 inÃ§ yÃ¼ksekliÄŸinde olacaktÄ±r.
* xlim=(0, 10000): Bu, x-ekseninin sÄ±nÄ±rlarÄ±nÄ± 0 ile 10000 arasÄ±nda ayarlamak iÃ§in kullanÄ±lÄ±r.
##### .map(sns.kdeplot, "price", fill=True):
* map metodu, belirtilen fonksiyonu grid Ã¼zerindeki her bir subplot iÃ§in uygular.
 * sns.kdeplot: Kernel Density Estimate (KDE) plot fonksiyonu.
* "price": KDE plot'un x ekseninde kullanacaÄŸÄ± veri sÃ¼tunu.
* fill=True: KDE eÄŸrisi altÄ±nda kalan alanÄ± gÃ¶lgelendirir.
* .add_legend(): GrafiÄŸe bir legend (aÃ§Ä±klama) ekler. Bu, hangi rengin hangi "cut" kategorisine karÅŸÄ±lÄ±k geldiÄŸini gÃ¶sterir.

```python
(sns
 .FacetGrid(df,
           hue = "cut",
           height = 5,
           xlim = (0,10000))
 .map(sns.kdeplot,"price",fill = True)
 .add_legend()
);
# eski sÃ¼rÃ¼mde fill yerine shade vardÄ±
```
